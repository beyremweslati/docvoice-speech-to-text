```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Création du fichier JSON pour mapper le paire audio-text

```python
audio_dir = "../dataset/processed/audio_final"
transcript_dir = "../dataset/processed/transcripts"
output_json = "../dataset/processed/json_format/whisper_dataset.json"
data = []
```

```python
for filename in sorted(os.listdir(audio_dir)):
    if filename.endswith(".wav"):
        base_name = os.path.splitext(filename)[0]
        audio_path = os.path.join(audio_dir, filename)
        text_path = os.path.join(transcript_dir, f"{base_name}.txt")

        # Ensure transcript exists
        if os.path.exists(text_path):
            with open(text_path, "r", encoding="utf-8") as tf:
                transcript = tf.read().strip()
            audio_path = audio_path.replace("\\", "/")
            data.append({
                "audio": audio_path,
                "text": transcript
            })
        else:
            print(f"No transcript found for {filename}")
```

```python
with open(output_json, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
print("Dataset created")
```

```python
df = pd.DataFrame(data)
df.head()
```

# Histogramme de la Durée des Clips Audio

```python
from pydub import AudioSegment
def calculate_duration(file_path):
    try:
        audio = AudioSegment.from_file(file_path)
        return len(audio) / 1000.0 
    except Exception as e:
        print(f"Erreur lors du traitement de {file_path}: {e}")
        return None
```

```python
df['duration_seconds'] = df['audio'].apply(calculate_duration) # Appel du fonction calculate_duration pour chaque 'audio' du df comme paramètre.
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['duration_seconds'], color='gold', edgecolor='black')
plt.title('Distribution des Durées des Clips Audio', fontsize=14)
plt.xlabel('Durée (secondes)', fontsize=12)
plt.ylabel('Fréquence (Nombre de Clips)', fontsize=12)
plt.grid(axis='y')
plt.show()
# Elle confirme que la majorité des clips sont très courts ( entre 0.8 et 2 secondes) (Génération correcte)
# ce qui est la durée attendue pour des mots uniques et prouve que le processus de génération avec ElevenLabs a été correct.
# Il n'y a pas d'erreurs d'enregistrement majeures (outliers) comme des clips de 5, 10, ou 30 secondes
```

# Corrélation Durée Audio et Longueur des Caractères

```python
df['char_length'] = df['text'].apply(len)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df['char_length'],
    df['duration_seconds'],
    alpha=0.6,
    color='darkorange'
)

plt.title('Corrélation entre Longueur du Terme et Durée de Prononciation', fontsize=14)
plt.xlabel('Nombre de Caractères dans le Terme', fontsize=12)
plt.ylabel('Durée de l\'Audio (secondes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# La présence d'une corrélation positive claire est la preuve que toutes les paires audio-texte correspondent correctement (Génération correcte)
```

# Évaluation du modèle de base Whisper

```python
from tqdm.auto import tqdm
from jiwer import wer, cer
from scipy.io.wavfile import read
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from datasets import load_dataset, Audio
```

```python
dataset = load_dataset("json", data_files={"train":"../dataset/processed/json_format/train.json", "test":"../dataset/processed/json_format/test.json"})
train_dataset = dataset["train"]
test_dataset = dataset["test"]
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "openai/whisper-medium.en"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)

processor = AutoProcessor.from_pretrained(model_id)
```

```python
def predictData(dataset):
    pred_texts, ref_texts = [], []
    for sample in tqdm(dataset, desc="Predecting Speech"):
        audio_path = sample["audio"]
        sample_rate, waveform = read(audio_path)
        input_features = processor(waveform, sampling_rate=sample_rate, return_tensors="pt").input_features.to(torch.float16)
        with torch.no_grad():
            predicted_ids = model.generate(input_features.to("cuda"))[0]
        transcription = processor.decode(predicted_ids)
        pred_texts.append(processor.tokenizer._normalize(transcription))
        ref_texts.append(sample["text"].lower().strip())
    return pred_texts,ref_texts
```

```python
pred_texts, ref_texts = predictData(train_dataset)
```

```python
baseWer = wer(ref_texts, pred_texts) * 100
baseCer = cer(ref_texts, pred_texts) * 100
print(f"Whisper WER (train_dataset): {baseWer:.2f}%")
print(f"Whisper CER (train_dataset): {baseCer:.2f}%")
```

```python
pred_texts, ref_texts = predictData(test_dataset)
```

```python
baseWer = wer(ref_texts, pred_texts) * 100
baseCer = cer(ref_texts, pred_texts) * 100
print(f"Whisper WER (test_dataset): {baseWer:.2f}%")
print(f"Whisper CER (test_dataset): {baseCer:.2f}%")
```
