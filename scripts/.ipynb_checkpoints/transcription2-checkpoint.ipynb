```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
from peft import PeftModel

device = "cuda" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32
model_id = "openai/whisper-medium.en"

lora_dir = "./whisper-docvoice-lora/final_adapter" 

# Load the base Whisper model
base_model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id,
    torch_dtype=torch_dtype,
    low_cpu_mem_usage=True
)

processor = AutoProcessor.from_pretrained(model_id)

# Load + merge LoRA
model = PeftModel.from_pretrained(base_model, lora_dir)
model = model.merge_and_unload()   # Combine LoRA weights into the main model

model.to(device)
```

```python
import sounddevice as sd
import numpy as np

SAMPLE_RATE = 16000
DURATION = 5

def record_audio():
    print("Recording")
    audio = sd.rec(int(DURATION * SAMPLE_RATE), samplerate=SAMPLE_RATE, channels=1, dtype="float32")
    sd.wait()
    return audio.squeeze()

def transcribe(audio_np):
    print("Transcribing, please wait.")
    inputs = processor(
        audio_np,
        sampling_rate=SAMPLE_RATE,
        return_tensors="pt"
    ).to(device)

    inputs["input_features"] = inputs["input_features"].to(dtype=torch_dtype)

    with torch.no_grad():
        predicted_ids = model.generate(
            inputs["input_features"],
            max_length=448
        )

    transcription = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
    return transcription

full_transcription = ""

while True:
    audio = record_audio()
    text = transcribe(audio)

    # Append transcription
    full_transcription += " " + text.strip()

    print("\nðŸ“œ Full transcription so far:")
    print(full_transcription)
```

```python
import sounddevice as sd
import numpy as np
import queue
import torch

q = queue.Queue()

SAMPLE_RATE = 16000
CHUNK = 16000 // 2     # chunk = 0.5 sec
WINDOW = 16000 * 3     # window = 3 sec
OVERLAP = 16000 * 2    # 2 sec overlap

audio_buffer = np.zeros(WINDOW, dtype=np.float32)

def callback(indata, frames, time, status):
    q.put(indata.copy())

stream = sd.InputStream(
    blocksize=CHUNK,      # <--- FIXED CHUNK SIZE
    callback=callback,
    channels=1,
    samplerate=SAMPLE_RATE
)

print("ðŸŽ¤ Streaming... speak!")

with stream:
    while True:
        chunk = q.get().squeeze()

        # Shift buffer (sliding window)
        audio_buffer[:-CHUNK] = audio_buffer[CHUNK:]
        audio_buffer[-CHUNK:] = chunk

        # Only process every 0.5 sec
        inputs = processor(
            audio_buffer,
            sampling_rate=SAMPLE_RATE,
            return_tensors="pt"
        ).to(device)

        inputs["input_features"] = inputs["input_features"].to(dtype=torch_dtype)

        with torch.no_grad():
            ids = model.generate(inputs["input_features"], max_length=448)

        text = processor.batch_decode(ids, skip_special_tokens=True)[0]

        print("ðŸ“", text)
```
